```python
import re
with open("./a.txt", "r", encoding = "utf-8") as file:
    txt = file.read()
#извлекаем файл а и присваиваем ее к переменной тхт
```

1

```python
print(re.search(r"ab*", txt).group())
#group выводит только 1 параметр в котором нащ стринг
```

2

```python
print(re.search(r"ab{2,3}", txt).group())
#group выводит только 1 параметр в котором нащ стринг
```

3

```python
re.findall(r"[a-z]+_[a-z]+", txt)
```

4

```python
re.findall(r"[A-Z][a-z]+", txt)
```

5

```python
re.findall(r"a.*b", txt)
```

6

```python
print(re.sub(r"[ ,.]", ":", txt))
```
